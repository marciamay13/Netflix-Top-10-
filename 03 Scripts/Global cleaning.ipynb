# 04. Consistency Checks

```python
# finding mixed-type columns
for col in df_global.columns.tolist():
  weird = (df_global[[col]].map(type) != df_global[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_global[weird]) > 0:
    print (col)
```

```python
#Changing the mixed type data column 'season_title' to 'str'
df_global['season_title'] = df_global['season_title'].astype('str')
```

```python
#dropping the episode launch column as it is not necessary for this analysis
df_global.drop(columns = ['episode_launch_details'])
```
